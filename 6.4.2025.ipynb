```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
#שיעורי בית למצגת 22 שקף 25-26
products_df = pd.DataFrame({
    'Product_ID': [1, 2, 3],
    'Product_Name': ['iPhone 12', 'Samsung S21', 'Pixel 5'],
    'Product_Price': [999, 899, 699],
})

sales_df = pd.DataFrame({
    'Sale_ID': [1, 2, 3, 4, 5, 6, 7],
    'Sale_Date': pd.to_datetime(['2021-01-01', '2021-02-01', '2021-02-15', '2020-03-01', '2020-04-01',
    '2020-02-15', '2020-01-01']),
    'Customer_ID': [1, 2, 3, 2, 3, 4, 4],
    'Product_ID': [1, 1, 2, 2, 3, 3, 1],
})

customers_df = pd.DataFrame({
    'Customer_ID': [1, 2, 3, 4],
    'Customer_Name': ['John Doe', 'Jane Doe', 'Jim Brown', 'Jake Smith'],
    'Customer_Age': [25, 30, 35, 20],
    'Customer_Email': ['john@doe.com', 'jane@doe.com', 'jim@brown.com', 'jake@smith.com'],
})
```

```python
sales_customers = pd.merge(sales_df, customers_df, on='Customer_ID')

full_df = pd.merge(sales_customers, products_df, on='Product_ID')
full_df
```

```python
#Question num 1
result = full_df.groupby('Product_Name').agg({
    'Customer_ID': lambda x: len(set(x)),
    'Customer_Age': 'mean'
})

result.rename(columns={
    'Customer_ID': 'Customer_Count',
    'Customer_Age': 'Average_Age'
})
result
```

```python
#question num 2
full_df['Sale_Date'] = pd.to_datetime(full_df['Sale_Date'])
full_df['Sale_Year'] = full_df['Sale_Date'].dt.year
sales_per_year = full_df.groupby(['Sale_Year', 'Product_Name']).size().reset_index(name='Sales_Count')

sales_per_year
```
